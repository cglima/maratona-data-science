**Strings**

Uma string é uma sequencia imutável de caracteres, podendo ser caracteres alfanuméricos ou caracteres especiais.

```
nome = "Ciana"
print(type(nome))
```

```
# F string
print(f"Oie, eu sou a {nome}")
```

```
# Formatando com F string
x = 1.61803399
print(f"O número de ouro é {x:.2f}")
```

```
# concatenação
a = "chat"
b = "bot"
print(a + b)
```

```
# atribuição de novas strings
c = "ro"
print(c + b)
```

```
# slicing/fatiamento
frase = "Esta é uma string"
print(frase[0])
print(frase[8])
```

```
# string[começo:fim:passos]
print(frase[:4])
print(frase[::2])
```

```
# atributos do objeto str
dir(str)
```

```
# help
help(str)
```
