**Exercícios**

1. Crie um algoritmo que solicite o nome do usuário e printe na sequência: "Bem vindo, <nome do usuário>"

2. Crie um algoritmo que solicite do usuário nome e idade, printando na tela as informações organizadas em 1 frase.

3. Crie um algoritmo que solicite uma palavra e verifique se é um palíndromo, printando na tela o resultado.

**Exercício 1**

```
nome = input("Qual seu nome?")
print(f"Bem vindo(a), {nome}")
```

**Exercício 2**

```
nome = input("Qual seu nome?")
idade = int(input("Qual a sua idade?"))
print(f"{nome} tem {idade} anos")
```

**Exercício 3**

```
entrada = input("Digite uma palavra:")
reverso = entrada[::-1]

if (entrada == reverso):
  print(f"A palavra {reverso} é um palíndromo")
else:
  print(f"A palavra {reverso} não é um palíndromo")
```
