**Exercícios**

1. Crie um algoritmo que solicite do usuário um número e calcule se ele é POSITIVO ou NEGATIVO.

2. Crie um algoritmo que solicite peso e altura e como resultado print o IMC na tela

3. Crie um algoritmo que solicite do usuário e printe na tela PAR, quando número par e IMPAR, quando número ímpar.

EXERCÍCIO 1

```
a = int(input("Digite um número:"))
if a >= 0:
  print("O número é positivo")
else: 
  print("O número é negativo")
```

EXERCÍCIO 2

```
peso = float(input("Informe seu peso em kg:"))
altura = float(input("Informe sua altura em m:"))
imc = peso / (altura*altura)
print(f"O seu indice de massa corporal é {imc:.2f}")
```

**EXERCICIO 3**

```
numero = int(input("Digite um número:"))
if((numero % 2) == 0):
  print(f"{numero} é PAR")
else:
    print(f"{numero} é ÍMPAR")
```
