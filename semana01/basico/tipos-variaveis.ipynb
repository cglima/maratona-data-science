**Variáveis** são elementos básicos que armazenam os dados que precisamos no nosso algoritmo e, cada dado pode ser de um tipo diferente. Em poucas palavras, variável é um nome que faz referência a um valor.


#Tipos de dados

```
# inteiro
a = 1
print(type(a))
```

```
# float (ponto flutuante )
b = 3.14
print(type(b))
```

```
# string (cadeia de caracteres (texto e número))
c = "Python"
print(type(c))
```

```
# booleano
d = True
print(type(d))
```

# Convertendo os tipos

```
# float para inteiro
e = int(b)
print(e, type(e))
b_resto = b % 3 
print(b_resto)
```

```
# inteiro para float
f = float(a)
print(f, type(f))
```

```
# inteiro para string
g = str(a)
print(g, type(g))
```
