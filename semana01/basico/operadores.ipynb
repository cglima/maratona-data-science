**Operadores** são símbolos especiais que representam cálculos. Os operadores são $+$, $-$, $*$, / e $**$, que representam, adição, subtração, multiplicação, divisão e exponenciação, respectivamente. 

```
# soma números
print(5+1)
```

```
# soma strings (concatenação)
print ("chat" + "bot")
```

```
# subtração
print(150-50)
```

```
# multiplicação
print(5*3)
```

```
# divisão
print(100/3)
```

```
# exponenciação
print(2**3)
```

```
# parte inteira da divisão
print(100//3)
```

```
# resto da divisão
print(100%3)
```

# **Regras de Precedências (PEMDAS)**

Quando mais de um operador aparecer na mesma sentença, qual deve ser executado primeiro? 

P: de parênteses! tem a mais alta precedência e por isso, em casos onde houverem parênteses, estes cálculos serão realizados primeiro.

2 * (3+1)= 8

2 * 3 + 1 = 7

E: de exponenciação! segunda precedência mais alta

2 ** 1 + 1 = 3

MDAS: Multiplicação e Divisão são a terceira precedência mais alta, restando a adição e subtração em última posição na ordem de precedência

2 * 3 - 1 = 5 


# **Operadores Lógicos**

$ > $ | maior

$ < $ | menor

$ >= $ | maior ou igual

$ <= $ | menor ou igual

$ == $ | igual

$ != $ | diferente 

```
# maior > 
print (10 > 0 )
```

```
# menor < 
print (0 < 10)
```

```
# maior ou igual >= 
print(10 >= 10)
```

```
# menor ou igual <= 
print(10 <= 20)
```

```
# igual == 
print (10 == 10)
```

```
# diferente != 
print(10 != "10")
```
