# **If**

if- se 

else - senão

elif - senão se 

```
x = 10 
if x > 0 :
  print("Número positivo")
else:
  print("Número negativo")
```

# **Conectivos lógicos**

**and** - TRUE se ambas atendidas

**or** - TRUE se alguma atendida

**not** - negação


```
x = "janeiro"
if x == "janeiro" or x == "fevereiro" or x == "março":
  print("Trimestre 1")
elif x == "abril" or x == "maio" or x == "junho":
  print("Trimestre 2")
else:
  print("Trimestre 3")
```
